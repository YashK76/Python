**Create bar chart and line chart using Matplotlib**

---




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/householdtask3.csv')
```

```python
display(data.head(10))
```

```python
# Scatter plot with year aganist own
plt.scatter(data['year'],data['own'])

#Adding title to the plot
plt.title('Scatter Plot')

# Setting the X and Y label
plt.xlabel('year')
plt.ylabel('own')

#Showing the result
plt.show()
```

```python
# Line chart with year against income
plt.plot(data['year'])
plt.plot(data['income'])

#Adding title to the plot
plt.title('Line chart')

# Setting the X and Y label
plt.xlabel('year')
plt.ylabel('income')

#Showing the result
plt.show()
```

```python
# Bar plot
plt.bar(data['year'],data['income'])

#Adding title to the plot
plt.title('Bar Plot')

# Setting the X and Y label
plt.xlabel('year')
plt.ylabel('income')

#Showing the result
plt.show
```

```python
# Histogram
plt.hist(data['income'])

#Adding title to the plot
plt.title('Histogram')

# Setting the X and Y label
plt.xlabel('year')
plt.ylabel('income')

#Showing the result
plt.show
```
